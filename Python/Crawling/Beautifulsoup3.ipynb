```python
# requests get 함수로 변경 ★
from bs4 import BeautifulSoup
from urllib.request import urlopen
from urllib.parse import quote  # 한글 처리 함수
from urllib.error import HTTPError
from requests import get        # GET 방식 호출

def getbs(url):
    try:
        html = get(url).content
        bs = BeautifulSoup(html, 'html.parser')
    except HTTPError as e:
        print(e)
        return None
    else:
        return bs
```

```python
bs = getbs('https://search.daum.net/search?w=tot&DA=Z6T&q=달러환율')
print(type(bs))
tag = bs.select('em.txt_num')[0]
print(type(tag))
print(tag)
print('오늘의 환율:', tag.text)
```

### [02] 서울 날씨/온도의 수집
- https://www.weather.go.kr/weather/observation/currentweather.jsp
- 서울시의 관측위치, 기온 가져오기
- Network 상태가 안좋은 경우 내용이 출력 안될 수 있음 ★

```python
# https://www.weather.go.kr/weather/observation/currentweather.jsp
bs = getbs('https://www.weather.go.kr/weather/observation/currentweather.jsp')
print(bs)
```

```python
#weather_table > tbody > tr:nth-child(42) > td:nth-child(1) > a
tag = bs.select('#weather_table > tbody > tr:nth-child(42) > td:nth-child(1) > a')[0]
print(tag.text)
```

```python
tag = bs.select('#weather_table > tbody > tr:nth-child(42) > td:nth-child(2)')[0]
print(tag.text)
```

```python
# AWS 관측 장비 서울 주소
# body > p > table:nth-child(3) > tbody > tr > td > table > tbody > tr:nth-child(2) > td.text2
# https://www.weather.go.kr/weather/observation/aws_table_popup.jsp
bs = getbs('https://www.weather.go.kr/weather/observation/aws_table_popup.jsp')
print(bs)
```

```python
# 내용이 로딩되지 않음.
tag = bs.select('body > p > table:nth-child(3) > tbody > tr > td > table > tbody > tr:nth-child(2) > td.text2')
print(tag)
```
