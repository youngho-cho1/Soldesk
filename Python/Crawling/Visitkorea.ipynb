```python
import time          # sleep 지연
import os

import pandas as pd  # Excel과 비슷한 구조를 지원하는 데이터 분석 패키지

from bs4 import BeautifulSoup      # Scraping(Crawling) package
from urllib.request import urlopen # 인터넷 접속후 페이지 소스 loading
from urllib.parse import quote     # 한글 처리 함수
from urllib.error import HTTPError # HTTP Exception
from requests import get           # GET 방식 호출

from selenium import webdriver  # 크롤링 이벤트 발생

import cx_Oracle  # Oracle
from sqlalchemy import create_engine  # Pandas -> Oracle

def load(url, second): # 페이지 열기
    try:
        driver.get(url)
        time.sleep(second) # Web page loading을 기다림
    except HTTPError as e:
        print(e)
        return None
    
def getbs(): # bs 객체 생성
    try:
        bs = BeautifulSoup(driver.page_source, 'html.parser') 
    except HTTPError as e:
        print(e)
        return None
    else:
        return bs
```

```python
# import cell과 분리, 반복 실행시 창이 계속 열림으로 1회만 실행 권장
driver = webdriver.Chrome('C:/kd/chromedriver.exe')
# 같은 도메인이라도 데스크탑과 모바일에서의 접속에 따라 디자인이 다른
# 경우가 있음으로 screen width, height를 이용해 데스크탑에 접속한다는 표시
# 데스크탑 기준으로 크롤링 스크립트를 제작한 경우,
# 모바일에 태그 검색이 안되는 경우가 많음.
driver.set_window_size(1200, 800)
```

```python
# 여행 정보 검색
time.sleep(1) # 1초
bs = getbs()
tags = bs.select('div.area_txt')
print(len(tags))

for tag in tags:
    title = tag.select('div > a')[0] # 연속적으로 select 함수 사용 가능
    print(title.text)
    area = tag.select('div > p')[0]
    print(area.text)
    
    try: # 검색 태그 없는 경우 예외 메시지 출력 방지
        search = tag.select('p.tag_type')[0] 
    except Exception as e:
        # IndexError: list index out of range
        pass # 예외 메시지 출력 방지
    else:    
        print(search.text)
    print('-' * 60)        
```

```python
# 검색 결과를 list 변수에 저장
nos = []
titles = []
areas = []
searchs = []

time.sleep(1) # 1초
bs = getbs()
tags = bs.select('div.area_txt')

for no, tag in enumerate(tags): # 0~
    nos.append(no + 1)
    title = tag.select('div > a')[0]
    titles.append(title.text.strip()) # 공백 제거
    area = tag.select('div > p')[0]
    areas.append(area.text.strip())
    
    try:
        search = tag.select('p.tag_type')[0]
    except Exception as e:
        # IndexError: list index out of range
        # pass
        searchs.append('') # list index를 통일 시키기위해 빈 데이터 추가 ★
    else:    
        searchs.append(search.text.strip())
    
print(nos)    
```

```python
df = pd.DataFrame()
df['번호'] = nos
df['제목'] = titles
df['지역'] = areas
df['태그'] = searchs
df
```

```python
# dbms 저장: 분석된 결과를 최종적으로 실시간 웹서비스용으로 저장
# DataFrame -> Oracle
# Oracle 테이블 생성
conn = cx_Oracle.connect('kd/1234@127.0.0.1:1521/XE')
cursor = conn.cursor()

try:
    cursor.execute('DROP TABLE visit') # 존재하면 삭제됨. 
except Exception as e:
    pass

cursor.execute('''
CREATE TABLE visit (
  no    NUMBER(7)   NOT NULL,
  title VARCHAR(200) NOT NULL, 
  area VARCHAR(100)      NULL, 
  tag   VARCHAR(500)     NULL
)
''')

cursor.close()
conn.close()
```

```python
# 모든 레코드 삭제
conn = cx_Oracle.connect('kd/1234@127.0.0.1:1521/XE')
cursor = conn.cursor()

cursor.execute('''
  DELETE FROM visit
''')

cursor.close()
conn.close()
```

```python
# Pandas -> Oracle, 연속 실행 PK 에러 발생
# Dataframe 컬럼명과 Oracle 컬럼명 일치해야함.
df2 = df
df2.columns=['no', 'title', 'area', 'tag']
engine = create_engine('oracle+cx_oracle://kd:1234@localhost:1521/?service_name=XE', echo=False)

# name='visit': 테이블명, if_exists='append': 기존 테이블에 레코드 추가
df2.to_sql(name='visit', con=engine, if_exists='append', index=False)
# ERROR, X
# df2.to_sql(name='visit', con=engine, if_exists='replace', index=False) 
```
