### 하나의 이미지에서 얼굴 부분만 자르기(crop)
- 이미지의 테두리에 여백이 부족하면 여백을 추가하여 처리 할 것

```python
# 하나의 이미지의 얼굴을 검색하여 Crop
import cv2
import datetime
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# CPU 기반
face_cascade = cv2.CascadeClassifier('./haarcascade_frontalface_default.xml')

img = cv2.imread('./src/actor1.jpg')      # 1명의 배우
# img = cv2.imread('./src/actor2.jpg')    # 2명의 배우
# img = cv2.imread('./src/actor3.jpg')    # 3명의 배우
# img = cv2.imread('./src/actor4.jpg')    # 1명의 배우

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
faces = face_cascade.detectMultiScale(img, 1.3,5)
count = 0
for (x,y,w,h) in faces:
    cropped = img[y - int(h / 2.5):y + h + int(h / 4), x - int(w / 4):x + w + int(w / 4)]
#     cropped = img[y - 30:y + h + 20, x - 10:x + w + 10]
    # 이미지를 저장
    
    plt.imshow(cropped)  # 이미지 출력
    plt.axis("off") # off: 축을 출력하지 말것
    plt.show()     # 이미지 출력

    print('이미지 발견 px: ' + ' x:' + str(x) + ' y:' +str(y) + ' w:' + str(w) + ' h:' + str(h))
       
#     count += 1
#     now = datetime.datetime.now()
#     nowDatetime = now.strftime('%Y%m%d%H%M%S')
    

#     cv2.imwrite('./src_crop/actor1/' + nowDatetime + "_" + str(count) + ".jpg", cropped)
#     cv2.imwrite('./src_crop/actor4/' + nowDatetime + "_" + str(count) + ".jpg", cropped)
#     cv2.imwrite('./src_crop/actor7/' + nowDatetime + "_" + str(count) + ".jpg", cropped)

# Lock 걸림으로 주석처리
# cv2.imshow('Image view', img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```
