```python
# 5가지 분류: 25가지 집락 추출용 데이터 생성
# JAVA
# 0.096, 0.01, 0.01, 0.01, 0.01
grp1 = '0.96, 0.01, 0.01, 0.01, 0.01,'
grp2 = '0.01, 0.96, 0.01, 0.01, 0.01,'
grp3 = '0.01, 0.01, 0.96, 0.01, 0.01,'
grp4 = '0.01, 0.01, 0.01, 0.96, 0.01,'
grp5 = '0.01, 0.01, 0.01, 0.01, 0.96,'

java = (grp1 * 5)[:-1]
print(java)
```

```python
python = (grp2 * 5)[:-1]
print(python)
```

```python
total = [grp1, grp2, grp3, grp4, grp5]
ds = ''
label_cnt = 1;
for grp in total:
    for i in range(0, 10):
        line = (grp * 5)[:-1] + ',' + str(label_cnt)
        ds = ds + line + '\n'
    label_cnt = label_cnt + 1    
    
print(ds)
```

```python
with open('./book6.csv', 'w', encoding='utf-8') as fw:
    fw.write('header\n')
    fw.write(ds)
```

```python
with open('./book6.csv', 'r', encoding='utf-8') as fr:
    line_cnt = 0
    while True:
        line = fr.readline() # 1개의 라인을 읽어옴, "\n" 줄바꿈 기호도 읽어옴.
        if line.strip() == '':  # 파일의 마지막인지 검사
            break
        print(line, end='')
        line_cnt = line_cnt + 1
print('line_cnt:', line_cnt)        
```
