### 변수, 상수, 함수, 거듭 제곱과 제곱근, 다항식 함수

```python
import warnings
warnings.filterwarnings(action='ignore')

# matplotlib package의 그래프 결과를 현재 셀에 출력
%matplotlib inline   

import numpy as np
import matplotlib.pyplot as plt

a = 1.5  # 1.5: 상수
x = np.linspace(-1,1)  # x: 변수 -1부터 1의 범위, 기본 50개
print(x)
print(len(x))
y = a * x  # y: 변수, 단조 증가 함수 : f(x) = ax

plt.plot(x, y) # x axis, y asxis
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
# x가 0일경우 y는 무조건 0이 발생하는 문제 발생
# 공부를 안해도 0이 아닌 최소 섲억이 발생하는 사례가 있는 경우는 표현 못함.
# 공부시간 x, 성적: y y는 0이 아닌 경우가 다수 존재함.
```

```python
# 함수 f(x) = ax + b  a: 기울기, x: 데이터, b: y 절편(bias)
import numpy as np

def my_func(x):     # my_func라는 이름의 Python 함수로 수식을 구현
    return 3*x + 2  # 3x + 2, 1차 방정식, 직선의 방정식   x = 0이면 y는 2

x = 4  # 글로벌 변수이므로 위에 적은 인수 x와는 다른 변수
y = my_func(x)  # y = f(x)
print(y)
```

```python
x = np.linspace(-1,1)  # x: 변수 -1부터 1의 범위, 기본 50개
y = my_func(x)  # y: 변수, 단조 증가 함수

plt.plot(x, y)
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
```

```python
def score(lean, x):   # score라는 이름의 Python 함수로 수식을 구현
    return lean*x + 20  # 3x + 20, 1차 방정식, 직선의 방정식
```

```python
x = np.arange(0,21, 1)  # x: 변수 0부터 21의 범위
x
```

```python
y1 = score(0, x)  # y: 변수, 단조 증가 함수
y2 = score(1, x)
y3 = score(1.5, x)

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
# 단순 선형 회귀 분석
# A 개발자가 공부를 못하면 분석시험 성적을 20점 취득 예상
# A 개발자가 10시간 공부하면 분석시험 성적을 30점 취득 예상
```

```python
# 다항식을 그래프
def my_func(x):  
    return 3*x + 2*x + 1 # 단조 증가

x = np.linspace(-2, 2)
y = my_func(x)  # y = f(x)

plt.plot(x, y)
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
```

```python
# 2차원 다항식을 그래프
def my_func(x):  
    return 3*x**2 + 2*x + 1

x = np.linspace(-2, 2)
y = my_func(x)  # y = f(x)

plt.plot(x, y)
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
```

```python
# 3차원 다항식을 그래프
def my_func(x):  
    return 4*x**3 + 2*x**2 + x + 3

x = np.linspace(-2, 2)
y = my_func(x)  # y = f(x)

plt.plot(x, y)
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
```

```python
def my_func(x):  
    return x**3 - 2*x**2 - 3*x + 4

x = np.linspace(-2, 2)
y = my_func(x)  # y = f(x)

plt.plot(x, y)
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
```

```python
# sin함수와 cos함수를 그래프
# 유사도 문제 해결: NLP에서 긍정은 긍정인 문장과 각이 좁음
# 유사도 문제 해결: NLP에서 긍정과 부정이 등장하면 각이 180도로 방향이 다름
def my_sin(x):  
    return np.sin(x)  # sin(x)

def my_cos(x):
    return np.cos(x)  # cos(x)

x = np.linspace(-np.pi, np.pi)  # -π부터 π(라디안)까지
y_sin = my_sin(x)
y_cos = my_cos(x)

plt.plot(x, y_sin, label="sin")
plt.plot(x, y_cos, label="cos")
plt.legend()

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.show()
```

```python
# 탄젠트 함수: 시계열 처리, 순환(계절, 규칙 순환), 언어또한 순환함.
def my_tan(x):  
    return np.tan(x)  # tan(x)

x = np.linspace(-1.3, 1.3)  # -1.3부터 1.3(라디안)까지
y_tan = my_tan(x)

plt.plot(x, y_tan, label="tan")
plt.legend()

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.show()
```

수식$y=\sin x$、$y=\cos x$의 그래프를 여러 가지 x의 범위에 그리기

```python
def my_sin(x):  
    return np.sin(x)  # sin(x)

def my_cos(x):
    return np.cos(x)  # cos(x)

x = np.linspace(-2*np.pi, 2*np.pi)  # x의 범위를 지정
y_sin = my_sin(x)
y_cos = my_cos(x)

plt.plot(x, y_sin, label="sin")
plt.plot(x, y_cos, label="cos")
plt.legend()

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.show()
```
