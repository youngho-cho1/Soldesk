```python
data = {
    "김밥" : [35,50,15],
    "라면" : [62,26,12]
}
idx = ["좋음","보통", "싫음"]
df = pd.DataFrame(data, index=idx)
df
```

```python
df.sum()
```

```python
# 김밥의 평가 비율은?
df['김밥'].plot.pie()
plt.title('김밥의 평가 비율')
plt.show()
```

```python
df['김밥'].plot.pie(startangle=0)
plt.title('김밥의 평가 비율')
plt.show()
```

```python
df['김밥'].plot.pie(startangle=90)
plt.title('김밥의 평가 비율')
plt.show()
```

```python
df['김밥'].plot.pie(startangle=0, counterclock=False) # 시계 방향
plt.title('김밥의 평가 비율')
plt.show()
```

```python
df['김밥'].plot.pie(startangle=0, counterclock=True) # 시계 반대 방향
plt.title('김밥의 평가 비율')
plt.show()
```

```python
# labeldistance=0.5: 중심에서 떨어진 거리
df['김밥'].plot.pie(startangle=0, counterclock=True, labeldistance=0.5) 
plt.title('김밥의 평가 비율')
plt.show()
```

```python
# 상자수염 챠트, Candel chart, 이상치를 판단하는 기준
data = {
    "A" :[163.6, 172.6, 163.7, 167.1, 169.9, 173.9, 170.1, 166.2, 176.7, 165.4],
    "B" :[166.9, 172.7, 166.4, 173.4, 169.6, 171.8, 166.9, 168.2, 166.7, 169.8]
}
df = pd.DataFrame(data)
df.head()
```

```python
sns.boxplot(data=df)
plt.title('신장 분포')
plt.show()
```

```python
sns.boxplot(data=df, width=0.2)
plt.title('신장 분포')
plt.show()
# A 그룹이 분산이 큼, 학생들간에 키의 차이가 심하다.
# A 그룹의 학생들은 개인간 성적의 차이가 크다.
# A 그룹의 매출의 변동이 심하다.
```

```python
df.median() # 분산이 큰 경우 평균보다 중위수가 더욱 신뢰됨.
```

```python
# 산포도, 경향(트랜드) 파악, 추세선
data = {
    "신장" :[163.6, 172.6, 163.7, 169.9, 173.9, 166.2, 176.7, 165.4],
    "체중" :[50.5, 63.3, 48.5, 59.8, 69.8, 53.7, 70.3, 51.2]
}
df = pd.DataFrame(data)
df.head()
```

```python
# 키가 크면 체중도 일반적으로 증가한다.
df.plot.scatter(x='신장', y='체중', c='b') # c='b': color = blue
plt.title('신장과 체중의 관계')
plt.show()
```

```python
# figsize=(12,8): 인치 단위
df.plot.scatter(x="신장", y="체중", c="b", figsize=(12,6))

x=df.iloc[3]["신장"] # index 로 접근
y=df.iloc[3]["체중"]
plt.plot(x, y, c="r", marker="X", markersize=15)

plt.title("나의 위치")
plt.show()
```

```python
df.plot.scatter(x="신장", y="체중", c="b", figsize=(8,5))
plt.title("나의 위치")

x=df.iloc[3]["신장"]
y=df.iloc[3]["체중"]
plt.plot(x, y, c="r", marker="X", markersize=15)

plt.axvline(x=x, c="r", linestyle="--") 
plt.axhline(y=y, c="g", linestyle="--")

plt.show()
```

```python
df.plot.scatter(x="신장", y="체중", c="b", figsize=(8,5))
plt.title("나의 위치")

x=df.iloc[3]["신장"]
y=df.iloc[3]["체중"]
plt.plot(x, y, c="r", marker="X", markersize=15)

plt.axvline(x = x - 5, c="r", linestyle="--")
plt.axvline(x=x, c="r", linestyle="--")
plt.axvline(x = x + 5, c="r", linestyle="--")

plt.axhline(y = y - 5, c="g", linestyle="--")
plt.axhline(y=y, c="g", linestyle="--")
plt.axhline(y = y + 5, c="g", linestyle="--")

plt.show()
```
