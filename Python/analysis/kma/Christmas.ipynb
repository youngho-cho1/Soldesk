```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import rc

rc('font', family='Malgun Gothic')
plt.rcParams["font.size"] = 12         # 글자 크기
plt.rcParams['axes.unicode_minus'] = False # minus 부호는 unicode 적용시 한글이 깨짐으로 설정

%matplotlib inline
```

```python
# 서울 지역의 매년 8월의 최고 기온만 필터링하여 plot 출력
# 날짜,지점,평균기온(℃),최저기온(℃),최고기온(℃)
# 1907-10-01,108,13.5,7.9,20.7
f = open('./data/seoul.csv', 'r', encoding='utf-8') # r: read
data = csv.reader(f, delimiter=',') # delimiter: 구분자
next(data) # header

filtered = [] # 8월의 최고 기온 저장
for row in data: # 파일 읽기
    if row[-1] != '': # 결측치가 아닌 행을 대상으로 처리
        month = row[0].split('-')[1] # 날짜를 분할하여 2번째 값인 월을 추출
        if month == '08': # 8월인지 확인
            filtered.append(float(row[-1]))

print(filtered[0:50]) # 추출된 데이터 50건 출력

plt.figure(figsize=(12, 3))  # 단위: 인치
plt.plot(filtered, c='red', label='온도') # x: 카운트, y: 온도
plt.title('1907년 ~ 2022년 03월 최고 기온 변화')
plt.legend() # 범례
plt.grid(True)
plt.show()

f.close()
```

```python
# 서울 지역의 1980년부터 현재까지 8월의 최고 기온만 필터링하여 plot 출력
# 날짜,지점,평균기온(℃),최저기온(℃),최고기온(℃)
# 1907-10-01,108,13.5,7.9,20.7
f = open('./data/seoul.csv', 'r', encoding='utf-8') # r: read
data = csv.reader(f, delimiter=',') # delimiter: 구분자
next(data) # header

filtered = [] # 8월의 최고 기온 저장
for row in data:
    if row[-1] != '': # '최고 기온'이 결측치가 아닌 행을 처리
        year = row[0].split('-')[0] # 년도
        month = row[0].split('-')[1] # 월
        if year >= '1980' and month == '08': # 1980년이후 8월달
            filtered.append(float(row[-1]))

print(filtered[0:50])

plt.figure(figsize=(12, 3))  # 단위: 인치
plt.plot(filtered, c='red', label='온도')
plt.title('1980년 ~ 2019년 8월 최고 기온 변화')
plt.legend()
plt.grid(True)
plt.show()

f.close()
```

```python
# 매년 12월 25일 최고 기온만 필터링하여 plot 출력
# 날짜,지점,평균기온(℃),최저기온(℃),최고기온(℃): 1907-10-01,108,13.5,7.9,20.7
f = open('./data/seoul.csv', 'r', encoding='utf-8') # r: read
data = csv.reader(f, delimiter=',') # delimiter: 구분자
next(data) # header

filtered = [] # 8월의 최고 기온 저장
for row in data:
    if row[-1] != '':
        month = row[0].split('-')[1]
        day = row[0].split('-')[2]
        if month == '12' and day == '25':
            filtered.append(float(row[-1]))

print(filtered[0:50])

plt.figure(figsize=(12, 3))  # 단위: 인치
plt.plot(filtered, c='red', label='온도')
plt.title('매년 12월 25일 최고 기온')
plt.legend()
plt.grid(True)
plt.show()

f.close()
```

```python
# 2000년 이후 12월 25일 최저, 최고 기온만 필터링하여 plot 출력
# 날짜,지점,평균기온(℃),최저기온(℃),최고기온(℃): 1907-10-01,108,13.5,7.9,20.7
f = open('./data/seoul.csv', 'r', encoding='utf-8') # r: read
data = csv.reader(f, delimiter=',') # delimiter: 구분자
next(data) # header

low = []  # 최저 기온 저장
high = [] # 최고 기온 저장
for row in data:
    if row[-1] != '':
        year = row[0].split('-')[0]
        month = row[0].split('-')[1]
        day = row[0].split('-')[2]
        if year>= '2000' and month == '12' and day == '25':
            low.append(float(row[-2]))
            high.append(float(row[-1]))

plt.figure(figsize=(12, 3))  # 단위: 인치
plt.plot(low, label='최저 온도') # 색깔 자동 결정
plt.plot(high, label='최고 온도')
plt.title('2000년 이후 12월 25일 기온 분포')
plt.legend() # label 속성을 범례로 수집하여 출력
plt.grid(True)
plt.show()

f.close()
```

```python
# 2010~2021년 12월 25일 최저, 최고 기온만 필터링하여 plot 출력
# 날짜,지점,평균기온(℃),최저기온(℃),최고기온(℃): 1907-10-01,108,13.5,7.9,20.7
f = open('./data/seoul.csv', 'r', encoding='utf-8') # r: read
data = csv.reader(f, delimiter=',') # delimiter: 구분자
next(data) # header

low = []  # 최저 기온 저장
high = [] # 최고 기온 저장
for row in data:
    if row[-1] != '':
        year = row[0].split('-')[0]
        month = row[0].split('-')[1]
        day = row[0].split('-')[2]
        if year >= '2010' and year <= '2021':
            if month == '12' and day == '25':
                low.append(float(row[-2]))
                high.append(float(row[-1]))

print(high)

plt.figure(figsize=(12, 3))  # 단위: 인치
plt.plot(low, label='최저 온도')
plt.plot(high, label='최고 온도')
plt.title('2000년 이후 12월 25일 기온 분포')
plt.legend()
plt.grid(True)
plt.show()

f.close()
```
