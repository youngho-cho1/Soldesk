```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import rc

rc('font', family='Malgun Gothic')
plt.rcParams["font.size"] = 12         # 글자 크기
plt.rcParams['axes.unicode_minus'] = False # minus 부호는 unicode 적용시 한글이 깨짐으로 설정

%matplotlib inline
```

```python
# 기상관측된 데이터 전체를 대상으로하여 서울 지역의 최고 온도를
# histogram으로 출력할것, 구간갯수: 100.
f = open('./data/seoul.csv', 'r', encoding='utf-8') # r: read
data = csv.reader(f, delimiter=',') # delimiter: 구분자
next(data) # header

high = []
for row in data:
    if row[-1] != '':
        high.append(float(row[-1]))

plt.figure(figsize=(12, 4))  # 단위: 인치
# bins=100: 영역을 100개로 구분
plt.hist(high, bins=100, color='tomato', label='최고 온도')
plt.title('서울 지역의 최고 온도')
plt.xlabel('온도')
plt.ylabel('빈도수')
plt.legend()
plt.grid(True)
plt.show()
        
f.close()
```

```python
# 기상관측된 데이터중 1월 데이터를 대상으로하여 서울 지역의 최고 온도를
# histogram으로 출력할것, 구간갯수: 100.
f = open('./data/seoul.csv', 'r', encoding='utf-8') # r: read
data = csv.reader(f, delimiter=',') # delimiter: 구분자
next(data) # header

high = []
for row in data:
    if row[-1] != '':
        month = row[0].split('-')[1]
        if month == '01':
            high.append(float(row[-1]))

plt.figure(figsize=(12, 4))  # 단위: 인치
plt.hist(high, bins=100, color='lightseagreen', label='최고 온도')
plt.title('서울 지역의 1월 최고 온도')
plt.xlabel('온도')
plt.ylabel('빈도수')
plt.legend()
plt.grid(True)
plt.show()
        
f.close()
```

```python
# 기상관측된 데이터중 8월 데이터를 대상으로하여 서울 지역의 최고 온도를
# histogram으로 출력할것, 구간갯수: 100.
f = open('./data/seoul.csv', 'r', encoding='utf-8') # r: read
data = csv.reader(f, delimiter=',') # delimiter: 구분자
next(data) # header

high = []
for row in data:
    if row[-1] != '':
        month = row[0].split('-')[1]
        if month == '08':
            high.append(float(row[-1]))

plt.figure(figsize=(12, 4))  # 단위: 인치
plt.hist(high, bins=100, color='fuchsia', label='최고 온도')
plt.title('서울 지역의 8월 최고 온도')
plt.xlabel('온도')
plt.ylabel('빈도수')
plt.legend()
plt.grid(True)
plt.show()
        
f.close()
```

```python
# 기상관측된 데이터중 1월, 8월 데이터를 대상으로하여 서울 지역의 최고 온도를
# histogram으로 출력할것, 구간갯수: 100.
f = open('./data/seoul.csv', 'r', encoding='utf-8') # r: read
data = csv.reader(f, delimiter=',') # delimiter: 구분자
next(data) # header

january = []
august = []
for row in data:
    if row[-1] != '':
        month = row[0].split('-')[1]
        if month == '01':
            january.append(float(row[-1]))
        elif month == '08':
            august.append(float(row[-1]))

plt.figure(figsize=(12, 4))  # 단위: 인치
plt.hist(january, bins=100, color='lightseagreen', label='1월 최고 온도')
plt.hist(august, bins=100, color='fuchsia', label='8월 최고 온도')
plt.title('서울 지역의 1월, 8월 최고 온도')
plt.xlabel('온도')
plt.ylabel('빈도수')
plt.legend()
plt.grid(True)
plt.show()
        
f.close()
```

```python
# 서울 지역의 최고 기온을 boxplot으로 출력할것.
f = open('./data/seoul.csv', 'r', encoding='utf-8') # r: read
data = csv.reader(f, delimiter=',') # delimiter: 구분자
next(data) # header

high = []
for row in data:
    if row[-1] != '':
        high.append(float(row[-1]))

plt.figure(figsize=(6, 4))  # 단위: 인치
plt.boxplot(high)
plt.title('서울 지역의 최고 온도')
plt.grid(True)
plt.show()
        
f.close()
# 데이터가 많아지면 이상치가 발견되지 않을 확률이 높음
```

```python
# 1월달의 최고기온을 대상으로 boxplot 출력
f = open('./data/seoul.csv', 'r', encoding='utf-8') # r: read
data = csv.reader(f, delimiter=',') # delimiter: 구분자
next(data) # header

high = []
for row in data:
    if row[-1] != '':
        month = row[0].split('-')[1]
        if month == '01':
            high.append(float(row[-1]))

plt.figure(figsize=(6, 4))  # 단위: 인치
plt.boxplot(high)
plt.title('서울 지역의 1월 최고 온도')
plt.grid(True)
plt.show()
        
f.close()
# 이상치가 다수 발생
```

```python
# 1월, 8월달의 최고기온을 대상으로 boxplot 출력
f = open('./data/seoul.csv', 'r', encoding='utf-8') # r: read
data = csv.reader(f, delimiter=',') # delimiter: 구분자
next(data) # header

january = []
august = []
for row in data:
    if row[-1] != '':
        month = row[0].split('-')[1]
        if month == '01':
            january.append(float(row[-1]))
        elif month == '08':
            august.append(float(row[-1]))

plt.figure(figsize=(12, 4))  # 단위: 인치
plt.boxplot([january, august]) # list를 이용한 2개 이상의 변수의 사용 
plt.title('서울 지역의 1월, 8월 최고 온도')
plt.grid(True)
plt.show()
        
f.close()
```
