### 결측치의 처리

```python
X.isnull().head()
```

```python
X.isnull()
```

```python
X.isnull().sum()
```

```python
X_cyl_mean = X['cyl'].mean() # cyl 변수의 평균
print(X_cyl_mean)
```

```python
# cyl 변수가 null인경우 평균으로 대체함
X['cyl'] = X['cyl'].fillna(X_cyl_mean)
print(X.isnull().sum())
print(X['cyl'])
```

```python
print(X['qsec'].median())
X['qsec'] = X['qsec'].fillna(X['qsec'].median())
print(X.isnull().sum())
print(X['qsec'])
```
