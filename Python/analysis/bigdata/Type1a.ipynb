```python
# 2. 분석 요쳥 사항
# 주어진 데이터의 첫 번째 행부터 순서대로 80%까지의 데이터를 훈련 데이터로 추출 후
# train 데이터에서 'total_bedrooms' 변수의 결측값(NaN)을
# 'total_bedrooms' 변수의 중앙값으로 대체하고,
# 대체전의 'total_bedrooms' 변수의 표준편차 값과
# 대체 후의 'total_bedrooms'변수의 표준편차 값을 산출한다. 
# 변환 이전과 이후의 표준편차 차이를 파악하여 제출해주세오. 
# Data: ./data/housing.csv  Answer: 1.975147
```

```python
df = pd.read_csv('./data/housing.csv')
df.describe()
```

```python
# 주어진 데이터의 첫 번째 행부터 순서대로 80%까지의 데이터를 훈련 데이터로 추출 후
print(df.shape)
per80 = int(df.shape[0] * 0.8)
print(per80)
train = df.iloc[0:per80,:].copy()
print(train.shape)
```

```python
# 대체전의 'total_bedrooms' 변수의 표준편차 값과
total_bedrooms_before = train['total_bedrooms'].std()
print('total_bedrooms_before:', total_bedrooms_before)

# train 데이터에서 'total_bedrooms' 변수의 결측값(NaN)을
# 'total_bedrooms' 변수의 중앙값으로 대체하고,
median = train['total_bedrooms'].median()
# print(median)
train['total_bedrooms'].fillna(median, inplace=True)
# print(train['total_bedrooms'].isnull().sum())

# 대체 후의 'total_bedrooms'변수의 표준편차 값을 산출한다. 
total_bedrooms_after = train['total_bedrooms'].std()
print('total_bedrooms_after:', total_bedrooms_after)
```

```python
# 변환 이전과 이후의 표준편차 차이를 파악하여 제출해주세오. 
print(abs(total_bedrooms_after - total_bedrooms_before))
```

```python
# 3. 분석 요쳥 사항
# Insurance 데이터 세트에서 charges 항목에서 이상값의 합을 파악하여 제출해주세오.
# (이상값은 평균에서 1.5 표준편차 이상인 값) 
# Data: ./data/insurance.csv  Answer: 6421430
```

```python
df = pd.read_csv('./data/insurance.csv')
df.head()
```

```python
df.describe()
```

```python
mean = df['charges'].mean()
std = df['charges'].std()
lower = mean - (std * 1.5) # 극단 경계 하한
upper = mean + (std * 1.5) # 극단 경계 상한
condition = (df['charges'] < lower) | (df['charges'] > upper)
print(sum(condition)) # 이상치 156 개
print(df[condition].shape)
print(df[condition]['charges'].sum())
```

```python
# 5. 분석 요쳥 사항
# Salary의 이상값의 합을 파악하여 제출해주세오.
# (단, 이상값은 중위수에서 IQR의 2배를 초과하는 값으로 한다.) 
# Data: ./data/Hitters.csv  Answer: 21671.864
```

```python
df = pd.read_csv('./data/Hitters.csv')
df.head()
```

```python
desc = df.describe()
print(desc)
```

```python
# Salary의 이상값의 합을 파악하여 제출해주세오.
# (단, 이상값은 중위수에서 IQR의 2배를 초과하는 값으로 한다.)
median = df['Salary'].median()
# print(median)
iqr = desc.loc['75%', 'Salary'] - desc.loc['25%', 'Salary']
# print(iqr)
lower = median - (iqr * 2)
# print(lower)
upper = median + (iqr * 2)
# print(upper)
cdt = (df['Salary'] < lower) | (df['Salary'] > upper)
# print(cdt)
ds = df[cdt]
# print(ds.shape)
print(ds['Salary'].sum())
```

```python
# 6_1. 분석 요청 사항  
# Diamonds의 데이터를 무작위로 비복원 추출하여 80%의 훈련 데이터로 분리후
# 'price' 변수를 기준으로 내림차순 정렬하여,
# 'price' 변수 기준으로 상위 100개의 데이터에 대한 price 평균을 구하시오. 
# Data: ./data/diamonds.csv  Answer: 18623.84
```

```python
# Diamonds의 데이터중 처음부터 순차적으로 80%의 데이터를 훈련 데이터로 만든후
# print(df.shape)
total_cnt = df.shape[0]
print(total_cnt)

cnt = int(total_cnt * 0.8)
print(cnt)

np.random.seed(0)
idx = np.random.choice(np.arange(0, total_cnt), size=cnt, replace=False)
print(idx)

ds = df.iloc[idx, :].copy()
print(ds.shape)

# print(ds.head())
# 'price' 변수를 기준으로 내림차순 정렬하여,
ds.sort_values(by='price', ascending=False, inplace=True)
# print(ds.head(10))
# 'price' 변수 기준으로 상위 100개의 데이터에 대한 price 평균을 구하시오. 
dt = ds.iloc[0:100, :]
# print(dt.shape)
print(dt['price'].mean())
```
