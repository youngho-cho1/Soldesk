#### 2. 데이터 수집이 정상적인지 확인하기위해 결측치가 포함된 관측치를 파악하여 제출할 것

```python
# 변수들마다 결측치의 갯수를 출력할 것
print(df.isnull().sum())
```

```python
# RM이 결측치인 관측치의 index를 파악할 것
# condition = df['RM'] == np.nan  --> X
# condition =(df['RM'] == 'NaN') --> X
# condition = (df['RM'] == None) --> X
condition = df['RM'].isnull()
print(sum(condition))
print(df[condition])
```

```python
# RM이 결측치인 관측치의 index
rm_index = df[condition].index
rm_index
```

```python
# RM이 결측치인 관측치의 index를 이용하여 'RM', 'MEDV'를 출력할 것 
df_rm = df.loc[df[condition].index, ['RM', 'MEDV']]
df_rm
```

```python
df_rm.sort_values(by=['MEDV'], ascending=True)
# boston 집값의 결측치를 분석한 결과 중간 가격대의 데이터가 누락이 많이 
# 발생되었음으로 추가적인 데이터 수집이나, 대체값으로 설정할 필요가 있음
```

```python
#누락된 데이터는 제외하고 집값(MEDV)의 평균과 표준편차를 산출할 것.
# 집값(MEDV)의 평균과 표준편차를 산출할 것.
condition = df['RM'].isnull()
df1 = df[-condition] #True -> False, False -> True
#df1.isnull().sum()
print(round(df1['MEDV'].mean(), 4))
print(round(df1['MEDV'].std(), 4))
```

```python
# 결측치가 있는 관측치를 무시하고 집값(MEDV)의 평균과 표준편차를 산출할 것.
print(round(df['MEDV'].mean(), 4))
print(round(df['MEDV'].std(), 4))
```

```python
# RM 변수의 결측치를 제외하고 평균과 표준편차를 산출할 것.
df2 = df.copy()
df2.dropna(inplace = True) # df2 바로 적용
#df2.isnull().sum()
print(df.shape)
print(df2.shape)
print('mean: ',df2['RM'].mean()) # 자동으로 결측치를 제외하고 처리함.
print('std: ',df2['RM'].std()) # 자동으로 결측치를 제외하고 처리함.
```

```python
# RM 변수의 결측치를 중위수로 대체하고 평균과 표준편차를 산출할 것.
med = df['RM'].median()
print(med)
df3 = df.copy()
df3['RM'].fillna(med, inplace=True)
print('mean: ',df3['RM'].mean()) # 평균이 중앙값으로 미세하게 이동함
print('std: ',df3['RM'].std()) # 중앙값이 추가되어 표준편차가 감소함.
```

#### 3. 부동산 투기로 인하여집값을 조사한 결과가 비정상적인 집값이 존재하는지 파악하여 출력할 것
- 이상치 하한: 평균 - (표준편차 *1.5)
- 이상치 상한: 평균 + (표준편차 *1.5)

```python
lower_limit = df['MEDV'].mean() - (df['MEDV'].std() * 1.5)
print('lower_limit: ', lower_limit)
upper_limit = df['MEDV'].mean() + (df['MEDV'].std() * 1.5)
print('upper_limit: ', upper_limit)
```

```python
condition = (df['MEDV'] < lower_limit) | (df['MEDV'] > upper_limit)
df[condition][['TAX', 'MEDV']].sort_values(by='MEDV',ascending=True)
# 저가 주택의 경우 복지 사각지대 존재할 수 있음으로 사회복지사 방문 등 필요
# 고가 주택의 경우 정상적인 세금 납부 여부 파악
```

#### 5.세금(TAX)을 기준으로 순위 구하기

```python
dfrank = df[['TAX']].head(10).sort_values(by=['TAX'], ascending=False)
dfrank
```

```python
dfrank['RANK'] = dfrank['TAX'].rank(ascending=False)
dfrank
```

```python
dfrank['RANK_AVERAGE'] = dfrank['TAX'].rank(method ='average',ascending=False)
dfrank
```

```python
#최소값(method='min') : 동점 관측치 그룹 내 최소 순위 부여
dfrank['RANK_MIN'] = dfrank['TAX'].rank(method ='min',ascending=False)
dfrank
```

```python
#최대값(method='max') : 동점 관측치 그룹 내 최고 순위 부여
dfrank['RANK_MAX'] = dfrank['TAX'].rank(method ='max',ascending=False)
dfrank
```

```python
#첫번째 값 (method='first') : 동점 관측치 중에서 데이터 상에서 먼저 나타나는 관측치부터 순위 부여
dfrank['RANK_FIRST'] = dfrank['TAX'].rank(method ='first',ascending=False)
dfrank
```

```python
#method='dense': 최소값('min')과 같은 방법으로 순위부여하나, 
#'min'과는 다르게 그룹 간 순위가 '1'씩 증가함
dfrank['RANK_DENSE'] = dfrank['TAX'].rank(method ='dense',ascending=False)
dfrank
```
