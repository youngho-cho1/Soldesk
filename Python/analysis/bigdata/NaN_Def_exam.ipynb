## NaN(Not a Number), fillna(), copy(), apply(), applymap(), map()
- map(): Series에 적용
- apply(): DataFrame, Series에 적용
- applymap(): DataFrame 전체에 적용, 함수안에서 함수 호출 가능

```python
import warnings
warnings.filterwarnings(action='ignore')
    
import numpy as np 
import pandas as pd
from pandas import Series, DataFrame
```

```python
df = pd.read_csv('./data/boston.csv')
df10 = df.head(10)
df10
```

```python
def nan(x):
    x = np.nan
    return x
df10.iloc[0:3, 0:4] = df10.iloc[0:3, 0:4].apply(nan)
df10
```

```python
def zero(col):
    x = 0
    return x
df10.iloc[0:3, 0:4] = df10.iloc[0:3, 0:4].apply(zero)
df10
```

```python
def five(x):
    x = 5
    return x
df10.iloc[0:3, 0:4] = df10.iloc[0:3, 0:4].applymap(five)
df10
```

```python
df10 = df10.applymap(zero)
df10.head()
```

```python
df10['TAX'] = 1000
df10.head()
```

```python
df10[['CRIM', 'ZN']] = df10[['CRIM', 'ZN']].apply(five)
df10.head()
```

```python
# TAX를 10$ 할인 적용할 것
def des(x):
    return x - (x*0.1)
df10['TAX'] = df10['TAX'].apply(des)
df10.head()
```

```python
# MEDV 컬럼 처음부터 5개의 관측치의 값을 5000으로 변경
df10.loc[0:5,"MEDV"] = 5000
df10
```

```python
#MEDV가 5000이 넘어가면 50%할인 그렇지 않으면 1000 할당
def proc(x):
    if x>=5000:
        x=x-(x*0.5)
    else:
        x=1000
    return x
df10['MEDV'] = df10['MEDV'].apply(proc) #inplace=True X
df10
```

## Lambda

```python
df = pd.read_csv('./data/boston.csv')
df10 = df.head(10)
df10
```

```python
#def nan(x):
#    x = np.nan
#    return x
df10.iloc[0:3, 0:4] = df10.iloc[0:3, 0:4].apply(lambda x: np.nan)
df10
```

```python
#def zero(col):
#    x = 0
#    return x
df10.iloc[0:3, 0:4] = df10.iloc[0:3, 0:4].apply(lambda x: 0)
df10
```

```python
#def five(x):
#    x = 5
#    return x
df10.iloc[0:3, 0:4] = df10.iloc[0:3, 0:4].applymap(lambda x: 5)
df10
```

```python
df10['TAX'] = df10['TAX'].apply(lambda x: 1000)
df10.head()
```

```python
df10['TAX'] = df10['TAX'].apply(lambda x: x - (x * 0.1))
df10.head()
```

```python
# DIS가 5미만이면 0, 초과이면 1을 할당
# lambda x:참 if x< 5 else 거짓
df10['DIS'] = df10['DIS'].apply(lambda x: 0 if x < 5 else 1)
df10.head()
```

```python
# INDUS가 5미만이면 0, 초과이면 1을 할당
# lambda x:참 if x< 5 else 거짓
#df10[['CRIM', 'INDUS']] = df10[['CRIM', 'INDUS']].apply(lambda x: 0 if x < 5 else 1)
df10[['CRIM', 'INDUS']] = df10[['CRIM', 'INDUS']].applymap(lambda x: 0 if x < 5 else 1)
df10.head()
```
