### 결측치의 처리

```python
X # 7번 관측치의 cyl 컬럼의 값이 누락되어 NaN으로 출력
```

```python
X.isnull() # NaN: True
```

```python
X.isnull().sum() # True:1, False: 0
```

```python
X_cyl_mean = X['cyl'].mean() # 결측치를 대체할 cyl 변수의 평균 산출
print(X_cyl_mean)
```

```python
# cyl 변수가 null인경우 평균으로 대체함
X['cyl'] = X['cyl'].fillna(X_cyl_mean)
print(X.isnull().sum())
print(X['cyl'])
```

```python
# 평균은 특이값(이상치)의 영향을 심하게 받는 단점이 존재함으로 중위수를
# 결측치의 대체 값으로 사용하는 경우도 많음.
print(X['qsec'].median()) # 데이터를 정렬하여 상위 50% 지점
X['qsec'] = X['qsec'].fillna(X['qsec'].median())
print(X.isnull().sum())
print(X['qsec'])
```

```python
# 잘못 수집된 데이터의 파악 및 처리
# 폐쇄형(객관식) 데이터 수집은 수집된 데이터가 비교적 정리가되어 있음
# 개방형(주관식) 데이터 수집은 직접 기입을 받음으로 전처리가 까다로움(권장안함)
# 객관식으로 변경이 어려움 데이터 수집
# 현재의 행복감은: 불행[0 - 1 - 2 - 3 - 4 - 5]행복
print(X['gear'].unique()) 
```
